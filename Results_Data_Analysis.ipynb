```python
import numpy as np


salet=np.zeros((7,2309))
crane=np.zeros((7,2309))
audio=np.zeros((7,2309))

for i in range(7):
    salet[i,:]=np.load('results/salet'+str(i+1)+'.npy')
    crane[i,:]=np.load('results/crane'+str(i+1)+'.npy')
    audio[i,:]=np.load('results/audio'+str(i+1)+'.npy')


```

```python
from matplotlib import pyplot as plt
import matplotlib 
matplotlib.rc('xtick', labelsize=18) 
matplotlib.rc('ytick', labelsize=18) 

plt.figure(figsize=(12,8))
plt.plot([1,2,3,4,5,6,7],np.mean(salet,axis=-1),label='SALET',linewidth=3)
plt.plot([1,2,3,4,5,6,7],np.mean(crane,axis=-1),label='CRANE',linewidth=3)
plt.plot([1,2,3,4,5,6,7],np.mean(audio,axis=-1),label='AUDIO',linewidth=3)
plt.legend(fontsize=20)
plt.ylabel('Average Number of Guesses',fontsize=20)
plt.xlabel('Cut Off',fontsize=20)
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12,8))
plt.plot([1,2,3,4,5,6,7],np.std(salet,axis=-1),label='SALET',linewidth=3)
plt.plot([1,2,3,4,5,6,7],np.std(crane,axis=-1),label='CRANE',linewidth=3)
plt.plot([1,2,3,4,5,6,7],np.std(audio,axis=-1),label='AUDIO',linewidth=3)
plt.legend(fontsize=20)
plt.ylabel('Standard Deviation of Guesses',fontsize=20)
plt.xlabel('Cut Off',fontsize=20)
plt.show()
```

```python
times=np.load('results/salet_times.npy')
```

```python
plt.figure(figsize=(12,8))
ts=salet[4,:]
bins=[1.5,2.5,3.5,4.5,5.5,6.5]
plt.hist(ts,histtype='step',bins=bins,linewidth=3)
plt.xlabel('Number of Guesses',fontsize=20)
plt.ylabel('Frequency',fontsize=20)
plt.plot([np.mean(ts),np.mean(ts)],[0,1600],color='red',linestyle='-',linewidth=3,label='mean')
plt.plot([np.mean(ts)+np.std(ts),np.mean(ts)+np.std(ts)],[0,1600],color='red',linestyle='--',linewidth=3,label='standard deviation')
plt.plot([np.mean(ts)-np.std(ts),np.mean(ts)-np.std(ts)],[0,1600],color='red',linestyle='--',linewidth=3)
plt.legend(fontsize=20)
plt.ylim(0,1300)
plt.show()
```

```python
plt.figure(figsize=(12,8))
ts=np.mean(times,axis=-1)
plt.hist(ts,histtype='step',bins=20,linewidth=3)
plt.xlabel('Time to Complete (Seconds)',fontsize=20)
plt.ylabel('Frequency',fontsize=20)
plt.plot([np.mean(ts),np.mean(ts)],[0,1000],color='red',linestyle='-',linewidth=3,label='mean')
plt.plot([np.mean(ts)+np.std(ts),np.mean(ts)+np.std(ts)],[0,1000],color='red',linestyle='--',linewidth=3,label='standard deviation')
plt.plot([np.mean(ts)-np.std(ts),np.mean(ts)-np.std(ts)],[0,1000],color='red',linestyle='--',linewidth=3)
plt.legend(fontsize=20)
plt.ylim(0,950)
plt.show()
```

```python
plt.figure(figsize=(12,8))
tpg=np.mean(times,axis=-1)/salet[4,:]
plt.hist(tpg,histtype='step',bins=20,linewidth=3)
plt.xlabel('Time per Guess (Seconds)',fontsize=20)
plt.ylabel('Frequency',fontsize=20)
plt.plot([np.mean(tpg),np.mean(tpg)],[0,1000],color='red',linestyle='-',linewidth=3,label='mean')
plt.plot([np.mean(tpg)+np.std(tpg),np.mean(tpg)+np.std(tpg)],[0,1000],color='red',linestyle='--',linewidth=3,label='standard deviation')
plt.plot([np.mean(tpg)-np.std(tpg),np.mean(tpg)-np.std(tpg)],[0,1000],color='red',linestyle='--',linewidth=3)
plt.legend(fontsize=20)
plt.ylim(0,800)
plt.show()
```
