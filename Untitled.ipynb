```python
import numpy as np


salet=np.zeros((7,2309))
crane=np.zeros((7,2309))
audio=np.zeros((7,2309))

for i in range(7):
    salet[i,:]=np.load('results/salet'+str(i+1)+'.npy')
    crane[i,:]=np.load('results/crane'+str(i+1)+'.npy')
    audio[i,:]=np.load('results/audio'+str(i+1)+'.npy')


```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12,8))
plt.plot([1,2,3,4,5,6,7],np.mean(salet,axis=-1),label='SALET',linewidth=3)
plt.plot([1,2,3,4,5,6,7],np.mean(crane,axis=-1),label='CRANE',linewidth=3)
plt.plot([1,2,3,4,5,6,7],np.mean(audio,axis=-1),label='AUDIO',linewidth=3)
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12,8))
plt.plot([1,2,3,4,5,6,7],np.std(salet,axis=-1),label='SALET',linewidth=3)
plt.plot([1,2,3,4,5,6,7],np.std(crane,axis=-1),label='CRANE',linewidth=3)
plt.plot([1,2,3,4,5,6,7],np.std(audio,axis=-1),label='AUDIO',linewidth=3)
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12,8))
plt.plot([1,2,3,4,5,6,7],np.max(salet,axis=-1),label='SALET',linewidth=3)
plt.plot([1,2,3,4,5,6,7],np.max(crane,axis=-1),label='CRANE',linewidth=3)
plt.plot([1,2,3,4,5,6,7],np.max(audio,axis=-1),label='AUDIO',linewidth=3)
plt.legend()
plt.show()
```
