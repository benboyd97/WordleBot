```python
import numpy as np

import sys

sys.path.append('../')
sys.path.append('./')
sys.path.append('wbot/')
sys.path.append('WordleBot/wbot/')

from wbot import read,play


#load list containing every possible wordle answer
plist=read.word_list('possible')

#load numpy array containing every allowed wordle guess 
glist=read.word_list('allowed')


all_tiles=[]


#for each guess and possible answer pair simulate tiles and save results to array
for i, g in enumerate(glist):
    g_outcomes=[]
    for j,p in enumerate(plist):
        g_outcomes+=[play.tiles(p,g)]
    
    all_tiles+=[g_outcomes]
    
    if i%100==0:
        print(i)



all_tiles=np.array(all_tiles)

#save all tiles as compressed file
np.savez_compressed('all_tiles.npz',data=all_tiles)
```

```python
import numpy as np

import sys

sys.path.append('../')
sys.path.append('./')
sys.path.append('wbot/')
sys.path.append('WordleBot/wbot/')

from wbot import read,play,solve

#load all tile combination for each guess word and true word pair
all_tiles=np.load('all_tiles.npz')['data']

#load list containing every possible wordle answer
plist=read.word_list('possible')

#load numpy array containing every allowed wordle guess 
glist=read.word_list('allowed')

#load the number of unique tile patterns given the starting guess is 'salet'
un_pat=np.unique(all_tiles[339,:,:],axis=0)


g_word=[]

#calculate the best second guess for possible pattern based on the maximum expected information
for pat in un_pat:
    

    tgrid,plist_=solve.reduce('audio',pat,all_tiles,np.array(plist),np.array(glist))

    info_grid=solve.info_grid(tgrid)

    guess=glist[np.argmax(info_grid)]
    
    g_word+=[guess]
    
    print(pat,guess)

#save unique patterns
np.save('audio_patterns.npy',un_pat)
#save best second guesses
np.save('audio_guess.npy',g_word,)
```
